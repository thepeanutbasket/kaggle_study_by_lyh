### 2.8 Fare

```python
fig, ax = plt.subplots(1, 1, figsize =(8, 8))
g = sns.distplot(df_train['Fare'], color = 'b', label = 'Skewness: {:.2f}'.format(df_train['Fare'].skew()), ax = ax)
g = g.legend(loc = 'best')
```

- You can see the distribution is not balances at all (=high skewness).
- I need to take the log on Fare.
- I will use lambda function to do this.

```python
df_test.loc[df_test.Fare.isnull(), 'Fare'] = df_test['Fare'].mean() # switching nan value to the mean value
df_train['Fare'] = df_train['Fare'].map(lambda i: np.log(i) if i > 0 else 0)
df_test['Fare'] = df_test['Fare'].map(lambda i: np.log(i) if i > 0 else 0)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (8, 8))
g = sns.distplot(df_train['Fare'], color = 'b', label = 'Skewness: {:.2f}'.format(df_train['Fare'].skew()), ax = ax)
g = g.legend(loc = 'best')
```

- After taking the log on Fare, the distribution looks more balanced than before.
